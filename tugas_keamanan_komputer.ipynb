```python
def rot13(text):
  result = ""
  for char in text:
    if char.isalpha():
      shift = 13
      if char.islower():
        result += chr((ord(char) - ord('a') + shift) % 26 + ord('a'))
      else:
        result += chr((ord(char) - ord('A') + shift) % 26 + ord('A'))
    else:
      result += char
  return result
```

```python
# contoh penggunaan
text = "Hello, World!"
encrypted = rot13(text)
print(f"Enkripsi: {encrypted}")
decrypted = rot13(encrypted)
print(f"Dekripsi: {decrypted}")
```
